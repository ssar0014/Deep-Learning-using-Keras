```python
from sklearn.model_selection import train_test_split
```

```python
X_test, X_train, y_test, y_train = train_test_split(iris['data'], iris['target'], random_state = 0)
```

```python
#make a pandas dataframe to visualise the feature dependencies 
iris_df = pd.DataFrame(X_train, columns = iris.feature_names)
```

```python
#iris_df
pd.plotting.scatter_matrix(iris_df, c = y_train, figsize=(15,15), marker = 'o', hist_kwds = {'bins':20}, s=60, alpha=0.8)
```

```python
#Building the KNN model
#import the K Neighbors Classifier class from the sklearn.neighbors module
#then initiate an object of the class so that it can be used for classification purposes
#number of neighbors taken for classification is 1

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
#train the model using the fit method
knn.fit(X_train, y_train)
```

```python
#make a new datapoint to check 
X_new = np.array([[3,1.9,0.2,1.2]])
```

```python
#make the model predict what this class new datapoint is in using the predict method
prediction = knn.predict(X_new)
```

```python
print("The new datapoint is belongs to: ", iris['target_names'][prediction])
```

```python
#test the accuracy of the predictions by using the predict method of the knn object and then using the score method
ypred = knn.predict(X_test)
knn.score(X_test,y_test)
```
